## Data Preprocessing

```python
df.isnull().sum()
```

```python
# caluclate the missing percentage in the data set 
missing_percentage=round((df.isnull().sum()/len(df)*100),2)
missing_percentage
```

```python
df['GENDER'].unique()
```

```python
df['GENDER'].value_counts()
```

Assuming that Majority of the customers are Females so we can fill the null values by female customers 

```python
# we can fill the null values by using fillna funtion by the values are Female customers
df['GENDER']=df['GENDER'].fillna('F')
df['GENDER'].isnull().sum()
```

Annual income Birthday count columns had the null values present in the data set so we can fill the missing values by using the mean values.

```python
mean=df['Annual_income'].mean()
mean
```

```python
df['Annual_income']=df['Annual_income'].fillna(mean)
```

```python
df['Annual_income'].isnull().sum()
```

```python
mean_birthday=df['Birthday_count'].mean()
mean_birthday
```

```python
df['Birthday_count']=df['Birthday_count'].fillna(mean_birthday)
df['Birthday_count'].isnull().sum()
```

```python
df['Type_Occupation'].value_counts()
```

Majority of the occupation is Laborers so we can fill the values by the value Laborers 

```python
mode=df['Type_Occupation'].mode()
mode[0]
```

```python
df['Type_Occupation']=df['Type_Occupation'].fillna(mode[0])
df['Type_Occupation'].isnull().sum()
```

we have to check the either null values present or not 

```python
df.isnull().sum()
```

There are no null values present in the data set so we can check the Duplicate entries in the data set and remove the duplicate entries in the data set 

```python
df.nunique()
```

```python
# checking the duplicate entries in the presenet in the data set 
df.duplicated().value_counts()
```

```python
## checking the duplicate withoun Ind_ID columns
duplicate=df.duplicated(subset=['GENDER','Car_Owner','Propert_Owner','CHILDREN','Annual_income','Type_Income','EDUCATION','Marital_status','Housing_type','Birthday_count','Employed_days','Mobile_phone','Work_Phone','Phone','EMAIL_ID','Type_Occupation','Family_Members','label'])
duplicate.value_counts()
```

```python
# we have to remove the duplicates 
df=df.drop_duplicates(subset=['GENDER','Car_Owner','Propert_Owner','CHILDREN','Annual_income','Type_Income','EDUCATION','Marital_status','Housing_type','Birthday_count','Employed_days','Mobile_phone','Work_Phone','Phone','EMAIL_ID','Type_Occupation','Family_Members','label'])
df.shape
```

We have to sucessfully remove the duplicate entries in the data set 

```python
# Rename the columns properly from the data set 
df=df.rename(columns={'GENDER':'Gender','Propert_Owner':'Property_Owner','CHILDREN':'Children','EDUCATION':'Education','EMAIL_ID':'Email_Id','label':'Label'})
df.head()
```

## Train Test split for Classification

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix, classification_report 
```

```python
categorical
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_columns=['Gender','Car_Owner','Property_Owner','Type_Income','Education','Marital_status','Housing_type','Type_Occupation']
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),categorical_columns)],remainder='passthrough')
```

## Logistic Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
lr=Pipeline(steps=[('preprocessor',ct),('model',LogisticRegression())])
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
```

```python
precison=precision_score(y_test,y_pred_lr)
precison
```

```python
print('Confusion_matrix \n' , confusion_matrix(y_test,y_pred_lr),"\n")
print('Accuracy_score:', round(accuracy_score(y_test,y_pred_lr)*100,2))
print('Precision_score:',round(precision_score(y_test,y_pred_lr)*100))
print('recall_score:',round(recall_score(y_test,y_pred_lr)*100,2))
print('f1_score:',round(f1_score(y_test,y_pred_lr)*100,2))
print('Train_score:', round(lr.score(X_train,y_train)*100,2))
print( classification_report (y_test,y_pred_lr))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=Pipeline(steps=[('preprocessor',ct),('model',DecisionTreeClassifier())])
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred_dt=dt.predict(X_test)
```

```python
print('Confusion_matrix \n' , confusion_matrix(y_test,y_pred_dt),"\n")
print('Accuracy_score:', round(accuracy_score(y_test,y_pred_dt)*100,2))
print('Precision_score:',round(precision_score(y_test,y_pred_dt)*100))
print('recall_score:',round(recall_score(y_test,y_pred_dt)*100,2))
print('f1_score:',round(f1_score(y_test,y_pred_dt)*100,2))
print('Train_score:', round(dt.score(X_train,y_train)*100,2))
print( classification_report (y_test,y_pred_dt))
```
